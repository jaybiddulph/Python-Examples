```python
#Practice

def fence(original, wrapper):
    output = wrapper+original+wrapper
    return output            # or you can just use the return wrapper+original+wrapper instead of output variable
print fence('name', '*')
print fence(1,3)   # look what happens when you put an integer
```

```python
from matplotlib import pyplot as plt
```

```python
x = [1,2,3]
y = [1,4,9]

plt.plot(x,y) # use the plot function of pyplot pass the variables
plt.show()
```

```python
x = [1,2,3]
y = [1,4,9]
z = [10,0,5]

plt.plot(x,y)
plt.plot(x,z)
plt.show()
```

```python
x = [1,2,3]
y = [1,4,9]
z = [10,0,5]

plt.plot(x,y, "-.", color = 'b', linewidth = 2.5) # parameters for graphing you can use "-." or "--"
plt.plot(x,z,  "--", color = 'r', linewidth = 2.5)
plt.show()
```

```python
x = [1,2,3]
y = [1,4,9]
z = [10,0,5]

plt.plot(x,y, "-", color = 'b', linewidth = 1.5) 
plt.plot(x,z,  "--", color = 'r', linewidth = 2.5)
plt.title("Toy Example") # graph title
plt.xlabel("x") # x axis labels
plt.ylabel("y and z") # y axis labels
plt.legend(['this is y','this is z']) # array put a legend
plt.show()
```

```python
# Scatter Plot
x = [1,2,3]
y = [1,4,9]

plt.scatter(x,y) #plot with default setting
plt.show()
```

```python
x = [1,2,3]
y = [1,4,9]

plt.scatter(x,y, s = 80, c = 'r', marker ='2') #plot with specific setttings
plt.show()
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
attr = iris.data
print iris.feature_names
print iris.feature_names[1]
```

```python
#Practice

plt.scatter(attr[:,0],attr[:,1], color='g', marker='D')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.show()
```

```python
target = iris.target
select_index = target == 0 # generate logical values based on whether Setosa
setosa = attr[select_index,:]
num_bins = 5 # number of bins
plt.hist(setosa[:,1], num_bins)  #give the frequency of observations in each bin

plt.xlabel('sepal width')
plt.ylabel('number of observations')
plt.title("Septosa Iris")
plt.show()
```
