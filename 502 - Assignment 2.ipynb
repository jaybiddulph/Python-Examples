```python
import pandas as pd
insurance = pd.read_csv("insurance.csv")
attr = insurance.values # convert the values in a data frame to a numpy array
print insurance.columns.values # print out headers (in the order)
from matplotlib import pyplot as plt
```

1. We will examine if bmi has an impact on the medical costs. Put the
bmi on the x-axis. The color of each point will be set according to
whether the patient is a smoker. Set the transparency to be 0.7. Be
sure to include the colorbar, and appropriate labels for x-axis and yaxis.
What business insights can you get?

```python
# Question 1
plt.scatter (attr[:,2], attr[:,-1], alpha = .7, c = attr[:,4]) 
plt.xlabel("BMI")
plt.ylabel("Cost")
plt.colorbar(ticks = [0,1])
plt.title ("BMI vs. Charges (non-smokers vs. smokers)")
plt.show()
```

2. We further compare the distribution of the medical costs of smokers
and that of non-smokers. Plot the distribution of medical costs of
smokers first. Then Plot the distribution of medical costs of nonsmokers
and set the transparency to 0.6. The number of bins is 10 for
both plots. Set appropriate labels and legends.

```python
#Question 2
index1 = attr[:,4] == 1
index2 = attr[:,4] == 0
plt.hist(attr[index1,-1], 10)
plt.hist(attr[index2,-1], 10, alpha =.6)
plt.xlabel('Health Care Cost')
plt.ylabel('# of Observations')
plt.legend(['smokers','non-smokers'])
plt.title('Medical Costs - smoker vs. non-smoker')
plt.show()
```

Question 3: We study whether age is an important factor by comparing the
distribution of medical costs of young people and that of elder people.
On the same plot, generate a histogram of medical costs of patients
younger than 40 years old, and then another histogram representing
the rest of the patients. Set the transparency of the second histogram
to 0.7. The number of bins is 15. Set appropriate labels and legends.
What can you conclude from this figure?

```python
index3 = attr[:,0] < 40
index4 = attr[:,0] >= 40
plt.hist(attr[index3,-1], 15)
plt.hist(attr[index4,-1], 15, alpha = .7)
plt.legend(['Age < 40','Age 40 and older'])
plt.xlabel('Healthcare Cost')
plt.ylabel('# of Observations')
plt.show()
```
