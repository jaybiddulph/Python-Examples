```python
select_index = target == 0
setosa = iris_attr[select_index,:]

num_bins = 5
plt.hist(setosa[:,1], num_bins)
plt.title("setosa")
plt.xlabel('sepal width')
plt.ylabel('number of observations')
plt.show()

select_index = target == 1
versicolor = iris_attr[select_index,:]

plt.hist(versicolor[:,1], num_bins)
plt.title("versicolor")
plt.xlabel('sepal width')
plt.ylabel('number of observations')
plt.show()
```

# Histograms are useful to determine outliers!!

```python
select_index = target == 0
setosa = iris_attr[select_index,:]

num_bins = 10
plt.hist(setosa[:,1], num_bins)  # histograms are useful to determine outliers!!
plt.title("setosa")
plt.xlabel('sepal width')
plt.ylabel('number of observations')
plt.show()
```

# Overlay Histogram

```python
target = iris.target
setosa = iris_attr[target == 0,:]
versicolor = iris_attr[target == 1,:]
virginica = iris_attr[target == 2,:]
num_bins = 15
plt.hist(setosa[:,1], num_bins, alpha = .75)
plt.hist(versicolor[:,1], num_bins, alpha = .75)
plt.hist(virginica[:,1], num_bins, alpha = .75)
plt.xlabel('sepal width')
plt.ylabel('number of observations')
plt.legend(['setosa', 'vesicolor', 'virginica'])
plt.show()
```

```python
import pandas as pd
daily = pd.read_csv("day.csv")  # use pandas to read csv file and assign it to daily
daily.head() #print daily first 5 rows
```

```python
attr = daily.values #convert daily.values to a numpy array called attr
print type(attr) #use type to confirm it is a numpy array
print daily.columns.values #print the header names of all the columns
```

```python
from matplotlib import pyplot as plt
plt.plot(attr[:,0],attr[:,9])
plt.xlabel('Day')
plt.ylabel('Check Outs')
plt.show()
```

This looks like summers have more checkouts.

```python
plt.scatter(attr[:,5],attr[:,7], color='#539cab', alpha = .7)
plt.ylabel('Casual Users')
plt.xlabel('Temp')

plt.show()
```

Looks like there is a correlation between checkouts and temperature

```python
plt.scatter(attr[:,5],attr[:,7], c = attr[:,3], alpha=.75)
plt.ylabel('Casual Users')
plt.xlabel('Temp')
plt.title('Casual Users vs. Workday, 1 = workday')
plt.colorbar(ticks = [0,1])
plt.show()
```

More casual users on non-workdays it seems

```python
bins = 10
plt.hist(attr[:,8], bins)
plt.title("Distribution of Registered Check Outs")
plt.xlabel('Registered Users')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.hist(attr[:,8], 10)
plt.hist(attr[:,7], 5, alpha = .8)
plt.title("Distribution of Registered Check Outs vs. Casual")
plt.xlabel('Check Outs')
plt.ylabel('Frequency')

plt.legend(['Registered','Casual'])
plt.show()
```

More Registered users checkout than casual
Casual checkouts is squewed and registered users has a normal (curved) distribution
